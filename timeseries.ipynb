```python
#importing libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading dataset
df=pd.read_csv("tested.csv")
```

```python
df.set_index('PassengerId', inplace=True)
```

```python
# Group every 50 passengers and compute survival rate per group
resampled = df['Survived'].groupby(df.index // 50).mean()
```

```python
# 10-passenger rolling mean for survival rate
df['rolling_survival'] = df['Survived'].rolling(window=10).mean()
```

```python
# Survival status of previous passenger for each row
df['Survived_lag1'] = df['Survived'].shift(1)
```

```python
plt.plot(df.index, df['rolling_survival'])
plt.xlabel('PassengerId')
plt.ylabel('Rolling Survival Rate')
plt.title('Rolling Average Survival Rate (Window=10)')
plt.show()
```
